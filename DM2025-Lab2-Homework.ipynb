Also, I incorporate a small portion of Gemini embeddings as additional features because it can capture contextual meaning and semantic similarity that TF-IDF alone can't represent.
Because there are limitations for free tier users, I only take very few embeddings as additional features and adjust their weight.

```python
from google import genai
from google.genai import types
SYSTEM_INSTRUCTION = (
    "You ara helpful assistant"
)
MAX_OUPTPUT_TOKENS = 65535
MODEL_NAME = "gemini-2.5-flash-lite"
SAFETY_SETTINGS = [
    types.SafetySetting(
        category="HARM_CATEGORY_HATE_SPEECH", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_DANGEROUS_CONTENT", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_SEXUALLY_EXPLICIT", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_HARASSMENT", threshold="OFF")
]
def prompt_gemini(
    input_prompt: list,
    temperature: float = 0.0,
    system_instruction: str = SYSTEM_INSTRUCTION,
    max_output_tokens: int = MAX_OUTPUT_TOKENS,
    client: genai.Client = client,
    model_name: str = MODEL_NAME,
    new_config: types.GenerateContentConfig = None,
    with_tools: bool = False,
    with_parts: bool = False,
    with_tokens_info: bool = False
  ):
```

```python
#gemini embedding
import os
from google import genai
import time
from google.api_core import exceptions
client = genai.Client(api_key=API_KEY)
def get_gemini_embedding(text: str, model: str="gemini-embedding-001",max_retry=5):
  INITIAL_WAIT_TIME = 5
  for attempt in range(max_retry):
    try:
      result = client.models.embed_content(model=model, contents=[text])
      time.sleep(0.6)
      return result.embeddings
    except exceptions.ResourceExhausted as e:
      print(f"Rate limit exceeded. Waiting to retry...Error:{e}")
      if attempt + 1 >= max_retry:
        print("Max retries reached")
        return None
      wait_time = INITIAL_WAIT_TIME * (2**attempt)
      time.sleep(wait_time)
      continue
    except Exception as e:
      print(f"An unexpected error occured: {e}")
      return None
```

```python
total_extractions = 300
train_ratio = 0.75
test_ratio = 0.25
n_train_to_sample = int(total_extractions * train_ratio)
n_test_to_sample = int(total_extractions * test_ratio)
column_name = 'text'
def stratified_sampled(post_df: pd.DataFrame, n_samples: int, stratify_col: str = 'emotion') -> pd.DataFrame:
  sampled_df = post_df.groupby(stratify_col, group_keys=False).apply(
      lambda x: x.sample(n=max(0, int(round(len(x) / len(post_df) * n_samples))))
  )
  current_samples = len(sampled_df)
  if current_samples < n_samples:
    remaining_indices = post_df.index.difference(sampled_df.index)
    additional_samples = post_df.loc[remaining_indices].sample(n=n_samples - current_samples, random_state=42)
    sampled_df = pd.concat([sampled_df, additional_samples])
  elif current_samples > n_samples:
    sampled_df = sampled_df.sample(n=n_samples, random_state=42)
  return sampled_df
```

```python
train_df_new = stratified_sampled(train_df, n_train_to_sample, 'emotion')
test_df_new = stratified_sampled(test_df, n_test_to_sample, 'emotion')
print("\nGenerating embeddings for the new training set...")
train_df_new['embeddings'] = train_df_new['text'].apply(get_gemini_embedding)
print("\nGenerating embeddings for the new testing set...")
test_df_new['embeddings'] = test_df_new['text'].apply(get_gemini_embedding)
```

```python
#using TFIDF into MNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, f1_score, classification_report
X_train = TFIDF_5000.transform(train_data['text'])
X_val = TFIDF_5000.transform(val_data['text'])
y_train = train_data['emotion']
y_val = val_data['emotion']
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
y_pred = mnb.predict(X_val)
acc = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average='weighted')
print("Validation Accuracy:", acc)
print("Validation F1-score:", f1)
print("\nClassification Report:\n", classification_report(y_val, y_pred))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_val)
acc = accuracy_score(y_val, pred_rf)
f1 = f1_score(y_val, pred_rf, average='weighted')
print("Validation Accuracy:", acc)
print("Validation F1-score:", f1)
print("\nClassification Report:\n", classification_report(y_val, pred_rf))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=3000)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_val)
acc = accuracy_score(y_val, pred_lr)
f1 = f1_score(y_val, pred_lr, average='weighted')
print("Validation Accuracy:", acc)
print("Validation F1-score:", f1)
print("\nClassification Report:\n", classification_report(y_val, pred_lr))
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
import numpy as np

k = 6
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(X_train_reduced)
train_cluster_labels = kmeans.labels_
val_cluster_labels = kmeans.predict(X_val_reduced)
ohe = OneHotEncoder(sparse_output=False)
train_cluster_ohe = ohe.fit_transform(train_cluster_labels.reshape(-1,1))
val_cluster_ohe = ohe.transform(val_cluster_labels.reshape(-1,1))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
X_train_final = np.hstack([X_train_combined, train_cluster_ohe])
X_val_final = np.hstack([X_val_combined, val_cluster_ohe])
y_train = train_data['emotion'].astype(str)
y_val = val_data['emotion'].astype(str)

lr5 = LogisticRegression(max_iter=3000)
lr5.fit(X_train_final, y_train)

pred_val_5 = lr5.predict(X_val_final)
print("Validation Accuracy:", accuracy_score(y_val, pred_val_5))
print("Validation F1-weighted:", f1_score(y_val, pred_val_5, average='weighted'))
```
