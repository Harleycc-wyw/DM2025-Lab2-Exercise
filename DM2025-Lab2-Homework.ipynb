```python
from google import genai
from google.genai import types
SYSTEM_INSTRUCTION = (
    "You ara helpful assistant"
)
MAX_OUPTPUT_TOKENS = 65535
MODEL_NAME = "gemini-2.5-flash-lite"
SAFETY_SETTINGS = [
    types.SafetySetting(
        category="HARM_CATEGORY_HATE_SPEECH", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_DANGEROUS_CONTENT", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_SEXUALLY_EXPLICIT", threshold="OFF"),
    types.SafetySetting(
        category="HARM_CATEGORY_HARASSMENT", threshold="OFF")
]
def prompt_gemini(
    input_prompt: list,
    temperature: float = 0.0,
    system_instruction: str = SYSTEM_INSTRUCTION,
    max_output_tokens: int = MAX_OUTPUT_TOKENS,
    client: genai.Client = client,
    model_name: str = MODEL_NAME,
    new_config: types.GenerateContentConfig = None,
    with_tools: bool = False,
    with_parts: bool = False,
    with_tokens_info: bool = False
  ):
```

```python
#gemini embedding
import os
from google import genai
import time
from google.api_core import exceptions
client = genai.Client(api_key=API_KEY)
def get_gemini_embedding(text: str, model: str="gemini-embedding-001",max_retry=5):
  INITIAL_WAIT_TIME = 5
  for attempt in range(max_retry):
    try:
      result = client.models.embed_content(model=model, contents=[text])
      time.sleep(0.6)
      return result.embeddings
    except exceptions.ResourceExhausted as e:
      print(f"Rate limit exceeded. Waiting to retry...Error:{e}")
      if attempt + 1 >= max_retry:
        print("Max retries reached")
        return None
      wait_time = INITIAL_WAIT_TIME * (2**attempt)
      time.sleep(wait_time)
      continue
    except Exception as e:
      print(f"An unexpected error occured: {e}")
      return None
```

```python
total_extractions = 300
train_ratio = 0.75
test_ratio = 0.25
n_train_to_sample = int(total_extractions * train_ratio)
n_test_to_sample = int(total_extractions * test_ratio)
column_name = 'text'
def stratified_sampled(post_df: pd.DataFrame, n_samples: int, stratify_col: str = 'emotion') -> pd.DataFrame:
  sampled_df = post_df.groupby(stratify_col, group_keys=False).apply(
      lambda x: x.sample(n=max(0, int(round(len(x) / len(post_df) * n_samples))))
  )
  current_samples = len(sampled_df)
  if current_samples < n_samples:
    remaining_indices = post_df.index.difference(sampled_df.index)
    additional_samples = post_df.loc[remaining_indices].sample(n=n_samples - current_samples, random_state=42)
    sampled_df = pd.concat([sampled_df, additional_samples])
  elif current_samples > n_samples:
    sampled_df = sampled_df.sample(n=n_samples, random_state=42)
  return sampled_df
```

```python
train_df_new = stratified_sampled(train_df, n_train_to_sample, 'emotion')
test_df_new = stratified_sampled(test_df, n_test_to_sample, 'emotion')
print("\nGenerating embeddings for the new training set...")
train_df_new['embeddings'] = train_df_new['text'].apply(get_gemini_embedding)
print("\nGenerating embeddings for the new testing set...")
test_df_new['embeddings'] = test_df_new['text'].apply(get_gemini_embedding)
```

```python
from google.genai import types
train_df_new['embeddings_values'] = train_df_new["embeddings"].apply(lambda row: list(types.ContentDict(row[0]).values())[0])
test_df_new['embeddings_values'] = test_df_new["embeddings"].apply(lambda row: list(types.ContentDict(row[0]).values())[0])
```

```python
# ===============================
# 1️⃣ Import 套件
# ===============================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score

# ===============================
# 2️⃣ train/val split
# ===============================
train_data, val_data = train_test_split(
    train_df,
    test_size=0.25,
    random_state=42,
    stratify=train_df['emotion']
)

# ===============================
# 3️⃣ TF-IDF 特徵
# ===============================
tfidf = TfidfVectorizer(max_features=5000)
tfidf.fit(train_df['text'])

X_train_tfidf = TFIDF_5000.transform(train_data['text']).toarray()
X_val_tfidf = TFIDF_5000.transform(val_data['text']).toarray()
X_test_tfidf = TFIDF_5000.transform(test_df['text']).toarray()

# ===============================
# 4️⃣ 加入 Gemini Embedding
# ===============================
embedding_dim = len(train_df_new['embeddings_values'].iloc[0])

# 建立全零矩陣
X_train_emb = np.zeros((len(train_data), embedding_dim))
X_val_emb = np.zeros((len(val_data), embedding_dim))
X_test_emb = np.zeros((len(test_df), embedding_dim))

# train_data 與 train_df_new 的 index 交集
train_new_index = train_data.index.intersection(train_df_new.index)
val_new_index = val_data.index.intersection(train_df_new.index)
test_new_index = test_df.index.intersection(test_df_new.index)

# 將 embeddings 放入對應 row
for idx in train_new_index:
    pos = train_data.index.get_loc(idx)
    X_train_emb[pos] = train_df_new.loc[idx, 'embeddings_values']

for idx in val_new_index:
    pos = val_data.index.get_loc(idx)
    X_val_emb[pos] = train_df_new.loc[idx, 'embeddings_values']

for idx in test_new_index:
    pos = test_df.index.get_loc(idx)
    X_test_emb[pos] = test_df_new.loc[idx, 'embeddings_values']

# 合併特徵
X_train_combined = np.hstack([X_train_tfidf, X_train_emb])
X_val_combined = np.hstack([X_val_tfidf, X_val_emb])
X_test_combined = np.hstack([X_test_tfidf, X_test_emb])

# ===============================
# 5️⃣ Label
# ===============================
y_train = train_data['emotion'].astype(str)
y_val = val_data['emotion'].astype(str)

# ===============================
# 6️⃣ 訓練 Logistic Regression
# ===============================
lr = LogisticRegression(max_iter=3000)
lr.fit(X_train_combined, y_train)

# ===============================
# 7️⃣ 驗證
# ===============================
pred_val = lr.predict(X_val_combined)
print("Validation Accuracy:", accuracy_score(y_val, pred_val))
print("Validation F1-weighted:", f1_score(y_val, pred_val, average='weighted'))


```
